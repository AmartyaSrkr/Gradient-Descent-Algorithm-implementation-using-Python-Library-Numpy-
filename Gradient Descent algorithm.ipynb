```python
# Implementing Gradient Descent for Linear Regression
# y_hat = wx + b (The linear function)
# loss = (y-y_hat)**2 / N

# First: Initializing some parameters
# Second: Creating the Gradient Descent function 
# Third: Iteratively make updates

# (1) Initializing some paramenters
import numpy as np
x = np.random.randn(10,1)
y = 2*x + np.random.rand()

# Parameters
w = 0.0
b = 0.0

# Hyperparameter
LearningRate_alpha = 0.01
print(x.shape[0])
# Creating the Gradient Descent function 
def GradientDescent(x,y,w,b,LearningRate_alpha):
    dldw = 0.0
    dldb = 0.0
    N = x.shape[0]
    # loss = (y-(wx+b))**2
    for x_i, y_i in zip(x,y):
        dl_dw += -2*x_i*(y_i-(w*x_i+b))
        dl_db += -2*(y_i-(w*x_i+b))

    w = w - LearningRate_alpha*(1/N)*dl_dw
    b = b - LearningRate_alpha*(1/N)*dl_db
    return w, b

# Iteratively makes updates of the parameteres 
for epoch in range(500):
    w, b = GradientDescent(x,y,w,b,LearningRate_alpha)
    y_hat = (w*x + b)
    loss = np.divide(np.sum((y-y_hat)**2, axis=0), x.shape[0])
    print(f'{epoch} results loss becomes {loss}, parameters w:{w}, b:{b}')
    # Run Gradient Descent 
    pass 
```
